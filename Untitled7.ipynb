Question 1:

What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.
-  Artificial Intelligence (AI):
AI is a broad field focused on creating machines that can perform tasks that require human intelligence, such as reasoning, learning, and problem-solving. Examples include chatbots, self-driving cars, and speech recognition systems.

- Machine Learning (ML):
ML is a subset of AI that enables systems to learn patterns from data and make predictions or decisions without explicit programming. Examples include spam detection, recommendation systems, and predictive analytics.

- Deep Learning (DL):
DL is a subset of ML that uses neural networks with multiple layers to automatically learn complex features from data. It is mainly used in image recognition, natural language processing, and speech recognition.

- Data Science:
Data Science combines statistical analysis, machine learning, and data visualization to extract insights and knowledge from data. It focuses on data cleaning, exploration, analysis, and storytelling for decision-making.

Question 2:

Explain overfitting and underfitting in ML. How can you detect and prevent them?

Answer:

- Overfitting:
A model that performs very well on training data but poorly on new data is said to be overfitted. It captures noise and irrelevant patterns. It can be detected when training accuracy is high, but validation accuracy is low.
Prevention: Use simpler models, apply regularization (L1/L2), cross-validation, dropout, and early stopping.

- Underfitting:
A model that performs poorly on both training and test data is underfitted. It fails to capture the underlying structure of data.
Prevention: Use a more complex model, train longer, reduce regularization, or add relevant features.

- Biasâ€“Variance Tradeoff:
Underfitting is due to high bias; overfitting is due to high variance. The goal is to balance both.


  
Question 3:

How would you handle missing values in a dataset? Explain at least three methods with examples.

Answer:

- Deletion:
Remove rows or columns with missing data if the missing portion is small and random.
Example: df.dropna() removes missing values.

- Imputation:
Replace missing numerical values with the mean or median, and categorical values with the mode.
Example: df['Age'].fillna(df['Age'].median(), inplace=True)

- Predictive Modeling:
Use regression or classification models to predict missing values based on other features.
Example: Using a regression model to estimate missing salary values based on experience and education.

Other methods include KNN imputation, interpolation for time series, and using a placeholder category for missing labels.


Question 4:

What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).

Answer:

An imbalanced dataset occurs when one class significantly outnumbers the others, such as 95% non-fraud and 5% fraud cases. This causes models to be biased toward the majority class.

- Technique 1: Resampling

Random Oversampling: Duplicate minority class samples.

Random Undersampling: Remove majority class samples.

SMOTE: Creates synthetic examples for the minority class.

- Technique 2: Class Weights
Assign higher weights to the minority class during training. Models like logistic regression or random forest can handle class_weight='balanced'.

Use precision, recall, F1-score, and ROC-AUC instead of accuracy to evaluate such models.

Question 5:

Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

Answer:

Feature scaling ensures that variables with large ranges do not dominate distance or gradient-based models. It helps algorithms like KNN, SVM, and logistic regression converge faster and perform better.

- Min-Max Scaling:
Scales values to a fixed range, usually [0,1].

- Standardization:
Centers data to mean 0 and standard deviation 1.

Min-Max is used when the model requires bounded input. Standardization is preferred when data has outliers or follows a normal distribution.

Question 6:

Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?

Answer:

- Label Encoding:
Assigns numeric values to each category. Example: Red=0, Green=1, Blue=2.
Best for ordinal data where order matters (e.g., low < medium < high).
Simple but may imply an order for non-ordinal data.

- One-Hot Encoding:
Creates separate binary columns for each category. Example: is_Red, is_Green, is_Blue.
Best for nominal data where no order exists.
Prevents misinterpretation but increases dimensionality.

Use label encoding for ordinal data and one-hot encoding for nominal data with few unique categories.


Question 7:

Google Play Store Dataset
Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?
 - Education, Medical, and Personalization apps often have the highest ratings because they serve focused user needs. Communication and Tools categories usually have lower ratings due to performance issues or ads.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/MasteriNeuron/datasets/master/googleplaystore.csv")
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df.dropna(subset=['Rating', 'Category'], inplace=True)
df = df[df['Category'].str.isalpha()]

category_rating = (
    df.groupby('Category')['Rating']
    .mean()
    .reset_index()
    .sort_values('Rating', ascending=False)
)

plt.figure(figsize=(8,6), dpi=150)
plt.barh(category_rating['Category'].head(15)[::-1],
         category_rating['Rating'].head(15)[::-1],
         color='#3b82f6')
plt.xlabel('Average Rating', fontsize=11)
plt.ylabel('Category', fontsize=11)
plt.title('Top 15 App Categories by Average Rating', fontsize=13, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Question 8:

Titanic Dataset
a) Compare the survival rates based on passenger class (Pclass).
b) Analyze how age affected survival.

Answer:
-
First-class passengers had the highest survival rate due to better access to lifeboats. Children had a higher survival rate than adults because of evacuation priority.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/MasteriNeuron/datasets/master/titanic.csv")
df['Survived'] = df['Survived'].astype(int)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

pclass_survival = df.groupby('Pclass')['Survived'].mean().reset_index()
print(pclass_survival)

df['Group'] = np.where(df['Age'] < 18, 'Child', 'Adult')
age_survival = df.groupby('Group')['Survived'].mean().reset_index()
print(age_survival)

plt.bar(pclass_survival['Pclass'].astype(str), pclass_survival['Survived']*100)
plt.xlabel('Pclass')
plt.ylabel('Survival Rate (%)')
plt.title('Survival by Passenger Class')
plt.show()

plt.bar(age_survival['Group'], age_survival['Survived']*100)
plt.xlabel('Age Group')
plt.ylabel('Survival Rate (%)')
plt.title('Survival by Age Group')
plt.show()
```

Question 9:

Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure?
b) Compare prices across airlines for the same route.

Answer:-
Prices remain stable for early bookings and rise exponentially as the departure date approaches. Budget airlines offer lower prices, while full-service airlines charge premium rates.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/MasteriNeuron/datasets/master/flight_price.csv")

cols = [c.lower().strip() for c in df.columns]
df.columns = cols

price_col = None
for c in cols:
    if 'price' in c:
        price_col = c
        break

if price_col is None:
    raise ValueError("No price-related column found in dataset.")

df[price_col] = pd.to_numeric(df[price_col], errors='coerce')

if 'days_left' not in df.columns:
    raise ValueError("Dataset missing 'days_left' column required for analysis.")

df['days_left'] = pd.to_numeric(df['days_left'], errors='coerce')
df.dropna(subset=[price_col, 'days_left'], inplace=True)

price_days = df.groupby('days_left')[price_col].median().reset_index().sort_values('days_left')
plt.figure(figsize=(8,5), dpi=150)
plt.plot(price_days['days_left'], price_days[price_col], marker='o')
plt.xlabel('Days Left')
plt.ylabel('Median Price')
plt.title('Flight Price vs Days Left')
plt.tight_layout()
plt.show()

if 'source' in df.columns and 'destination' in df.columns and 'airline' in df.columns:
    route = df[(df['source'].str.contains('delhi', case=False)) & (df['destination'].str.contains('mumbai', case=False))]
    if not route.empty:
        airline_price = route.groupby('airline')[price_col].median().reset_index().sort_values(price_col)
        plt.figure(figsize=(8,5), dpi=150)
        plt.bar(airline_price['airline'], airline_price[price_col], color='#3b82f6')
        plt.xticks(rotation=45)
        plt.xlabel('Airline')
        plt.ylabel('Median Price')
        plt.title('Delhiâ€“Mumbai Flight Prices by Airline')
        plt.tight_layout()
        plt.show()
```

Question 10:

HR Analytics Dataset
a) What factors most strongly correlate with employee attrition?
b) Are employees with more projects more likely to leave?

Answer:
- High overtime, low satisfaction, and low salary are the strongest predictors of attrition. Employees handling excessive projects show a higher tendency to leave due to workload stress.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv("https://raw.githubusercontent.com/MasteriNeuron/datasets/master/hr_analytics.csv")

cols = [c.lower().strip() for c in df.columns]
df.columns = cols

target_col = None
for c in cols:
    if 'attrition' in c or 'left' in c:
        target_col = c
        break

df[target_col] = df[target_col].map({'Yes':1,'No':0,'yes':1,'no':0,1:1,0:0})

num_df = df.select_dtypes(include=['int64','float64'])
corr = num_df.corr()[target_col].sort_values(ascending=False)
print(corr.head(10))

proj_col = None
for c in cols:
    if 'project' in c:
        proj_col = c
        break

if proj_col:
    proj = df.groupby(proj_col)[target_col].mean().reset_index()
    plt.figure(figsize=(7,4), dpi=150)
    plt.plot(proj[proj_col], proj[target_col]*100, marker='o', color='#3b82f6')
    plt.xlabel('Number of Projects')
    plt.ylabel('Attrition Rate (%)')
    plt.title('Attrition vs Number of Projects')
    plt.tight_layout()
    plt.show()

le = LabelEncoder()
for col in df.select_dtypes(include=['object']).columns:
    df[col] = le.fit_transform(df[col].astype(str))

X = df.drop(target_col, axis=1)
y = df[target_col].astype(int)

rf = RandomForestClassifier(random_state=42)
rf.fit(X, y)
imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print(imp.head(10))
```
